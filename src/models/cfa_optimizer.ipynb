```

# ONE-CLICK RUN CELL — run this cell, or use Kernel → Restart Kernel and Run All Cells

import os
import sys
import subprocess
import importlib
from pathlib import Path

print("=== CFA Optimizer Comparison: One-Click Run ===")

# -----------------------------------------------------------------------------
# 1) Install/check required packages BEFORE importing project modules
# -----------------------------------------------------------------------------
required_packages = [
    "pandas",
    "numpy",
    "scipy",
    "scikit-learn",
    "xgboost",
    "lightgbm",
    "matplotlib",
    "pydantic",
]

for package in required_packages:
    try:
        importlib.import_module(package.replace("scikit-learn", "sklearn"))
        print(f"Package ready: {package}")
    except ImportError:
        print(f"Installing missing package: {package}")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

print("\nAll required packages are installed/available.")

# -----------------------------------------------------------------------------
# 2) Set working folder to Downloads/Mod 5
# -----------------------------------------------------------------------------
PROJECT_FOLDER = Path.home() / "Downloads" / "Mod 5"

if not PROJECT_FOLDER.exists():
    raise FileNotFoundError(
        f"Could not find the folder: {PROJECT_FOLDER}\n"
        "Please confirm your folder name is exactly 'Mod 5' inside Downloads."
    )

os.chdir(PROJECT_FOLDER)
print("\nWorking folder set to:", Path.cwd())

print("\nFiles found in this folder:")
for item in sorted(os.listdir()):
    print("-", item)

# -----------------------------------------------------------------------------
# 3) Check project files
# -----------------------------------------------------------------------------
required_project_files = [
    "all_optimizers_combined_final.py",
    "problem_config.py",
    "data_module.py",
    "prediction_module.py",
]

missing_files = [f for f in required_project_files if not Path(f).exists()]
if missing_files:
    raise FileNotFoundError(
        "Missing required project files in Downloads/Mod 5:\n"
        + "\n".join(f"- {f}" for f in missing_files)
    )

if Path("cleaned_orders.csv").exists():
    CSV_PATH = "cleaned_orders.csv"
elif Path("orders.csv").exists():
    CSV_PATH = "orders.csv"
else:
    raise FileNotFoundError(
        "Could not find cleaned_orders.csv or orders.csv in Downloads/Mod 5."
    )

print(f"\nUsing data file: {CSV_PATH}")

# -----------------------------------------------------------------------------
# 4) Import and run full comparison
# -----------------------------------------------------------------------------
# Reload is helpful if you made changes to the .py file while the notebook is open.
import all_optimizers_combined_final
importlib.reload(all_optimizers_combined_final)
from all_optimizers_combined_final import run_full_comparison

comparison, results, best_result = run_full_comparison(
    csv_path=CSV_PATH,
    project_folder=PROJECT_FOLDER,
    n_restarts=10,
    max_iter=1500,
)

# -----------------------------------------------------------------------------
# 5) Display final results
# -----------------------------------------------------------------------------
print("\n=== FINAL RESULT ===")
print("Best optimizer:", best_result.method)
print("Best objective value:", f"${best_result.objective_value:,.2f}")
print("Total spend:", f"${best_result.total_spend:,.2f}")
print("Feasible:", best_result.feasible)

print("\nSaved output files in Downloads/Mod 5:")
print("- optimizer_comparison.csv")
print("- best_optimizer_order_plan.csv")
print("- all_optimizer_order_plans.csv")

comparison
```
